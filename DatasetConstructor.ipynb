```python
import os
from os import path
import pandas as pd
from pathlib import Path
```

```python
python_dataset = {
    "program_id": [],
    "source": [],
    "target": [],
    "diffculty": [],
}
java_dataset = {
    "program_id": [],
    "source": [],
    "target": [],
    "diffculty": [],
}
cpp_dataset = {
    "program_id": [],
    "source": [],
    "target": [],
    "diffculty": [],
}


for description in os.listdir("./descriptions/"):
    print(description[:-4])
    with open(path.join("descriptions", description), "r") as description_file,open(path.join("difficulties", description), "r") as difficulty_file:
        description_text = description_file.read()
        difficulty_text = difficulty_file.read()
        if description_text.strip() == "N/A":
            continue
        if os.path.exists(path.join("java", f"{description[:-4]}.java")):
            with open(path.join("java", f"{description[:-4]}.java"), "r") as java_file:
                wrapped_comment = f"/* {description_text} */\n"
                java = java_file.read()
                source = wrapped_comment + "\n".join(java.split("\n")[:2])

                java_dataset["program_id"].append(f"leetcode_java_{description[:-4]}")
                java_dataset["source"].append(source)
                java_dataset["target"].append(wrapped_comment + java)
                java_dataset["diffculty"].append(difficulty_text)
                pass
        if os.path.exists(path.join("py", f"{description[:-4]}.py")):
            with open(path.join("py", f"{description[:-4]}.py"), "r") as python_file:
                wrapped_comment = f'""" {description_text}\n """'
                python = python_file.read()
                source = wrapped_comment + "\n".join(python.split("\n")[:2])

                python_dataset["program_id"].append(
                    f"leetcode_python_{description[:-4]}"
                )
                python_dataset["source"].append(source)
                python_dataset["target"].append(wrapped_comment + python)
                python_dataset["diffculty"].append(difficulty_text)
                pass
        if os.path.exists(path.join("c++", f"{description[:-4]}.cpp")):
            with open(path.join("c++", f"{description[:-4]}.cpp"), "r") as cpp_file:
                wrapped_comment = f"/* {description_text} */\n"
                cpp = cpp_file.read()
                source = wrapped_comment + "\n".join(cpp.split("\n")[:3])

                cpp_dataset["program_id"].append(f"leetcode_cpp_{description[:-4]}")
                cpp_dataset["source"].append(source)
                cpp_dataset["target"].append(wrapped_comment + cpp)
                cpp_dataset["diffculty"].append(difficulty_text)
                pass
```

```python
python_dataset
```

```python
pd.DataFrame(python_dataset).to_csv("python_dataset.csv")
pd.DataFrame(java_dataset).to_csv("java_dataset.csv")
pd.DataFrame(cpp_dataset).to_csv("cpp_dataset.csv")
```
